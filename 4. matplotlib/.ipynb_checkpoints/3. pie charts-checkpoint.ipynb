```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
firms = ["Firm A", "Firm B", "Firm C", 
         "Firm D", "Firm E"]
market_shares = [10, 40, 30, 5, 15]
```

```python
Explode = [0,0,0,0.5,0]
```

```python
plt.pie(market_shares,explode=Explode,
       labels=firms,shadow=True,startangle=60)
plt.legend(title = "List of Firms",loc='best')
plt.axis('equal')
plt.show()

```
