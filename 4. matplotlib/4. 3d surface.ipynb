```python
def z_func (x,y):
    return (np.sin(np.sqrt(x**2 + y**2)))
```

```python
x = np.linspace(-15,15,100)
y = np.linspace(-15,15,100)
```

```python
X,Y = np.meshgrid(x,y)
Z = z_func(X,Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
x1 = np.arange(-5,5,0.1)
y1 = np.arange(-5,5,0.1)
```

```python
X1,Y1 = np.meshgrid(x1,y1)
Z1 = np.sin(X1) * np.cos(Y1)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X1,Y1,Z1,cmap='plasma')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X1,Y1,Z1,cmap='plasma')
ax.view_init(azim=0,elev=90)
plt.show()
# top view
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
figure = plt.figure(1, figsize = (50, 10))
subplot3d=plt.subplot(111,projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, 
                                 linewidth=0.1)
plt.show()
```
