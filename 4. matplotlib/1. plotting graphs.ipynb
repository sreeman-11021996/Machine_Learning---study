# Pie Chart

```python
areas = ['Marketing','Sales','Development','HR','Customer Support']
budjet = [2.5,4,10,1,6]
```

```python
plt.pie(budjet, labels=areas)
```

```python
plt.pie(budjet, labels=areas, autopct='%0.2f%%')
plt.show() 
# args in pie() : autopct -> to display percentages 
# c notation: %0.1f%% - for upto 1 decimal places
# plt.show() -> to avoid showing the object and text list at the start
```

```python
from numpy.random import rand
```

```python
# alpha gives the transperency of graph
```

```python
for color in ['red', 'green', 'blue']:
    n = 100
    x,y = rand(2,n)
    scale = 500.0 * rand(n) ** 5
    plt.scatter(x,y,s=scale,c=color,
               alpha=0.3,edgecolors='blue')
plt.grid(True)
plt.show()
```

# Linear

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233,456,770,120,222,444,200,300]
offline_price = [400,300,100,333,444,565,565,899]
chor_bazar = [100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price,color='red',marker='D',markersize=10,
         label='Online Price')
plt.plot(product_id,offline_price,color='yellow',marker='*',markersize=10,
         label='Offline Price')
plt.plot(product_id,chor_bazar,color='green',marker='+',markersize=10,
         label='Chor Bazaar')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.legend()
plt.title('Product Price Comparision.')
plt.show()
```

# Bar Graph

```python
year = [2012,2013,2014,2015,2016,2017,2018]
price = [65000,69000,75000,55000,52000,64000,81000]
```

```python
plt.bar(year,price,color='red')
plt.xlabel('Year')
plt.ylabel('Price(in INR)')
plt.title('Year by Year Price Comp.')
plt.show()
```

```python
plt.barh(year,price,color='green')
plt.ylabel('Year')
plt.xlabel('Price(in INR)')
plt.title('Year by Year Price Comp.')
plt.show()
```

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

# Box Plot

```python
import numpy as np
```

```python
# args 
#-> vert -> if true draws a vertical box plot
#-> patch_artist -> if true the box gets filled with a color
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.show
```
