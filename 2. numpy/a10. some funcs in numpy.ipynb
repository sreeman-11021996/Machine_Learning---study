```python
# solving a system of linear scalar equations
```

```python
coeffs = np.array([[2,6],[5,3]])
depvars = np.array([6,-9])
```

```python
solution = np.linalg.solve(coeffs,depvars)
```

```python
solution
```

```python
coeffs.dot(solution)
```
