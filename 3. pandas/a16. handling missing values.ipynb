```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data
```

# Dropna()

```python
data.dropna().shape    # drops all rows missing even one value/ how='any' - default
# 708 rows dropped
```

```python
data.dropna(how='all').shape     # drops only rows with all missing values
```

```python
data.dropna(axis=1).shape   # drops all cols. missing even one value 
# -> 2 cols. dropped
```

```python
data.dropna(axis=1,how='all').shape   # drops all cols. missing all values
```

```python
data.dropna(subset=['Cabin'])  # only drops rows in which cabin col. value is missing
```

```python
data.dropna(subset=['Cabin','Embarked'])
```

# fillna()

```python
data.fillna(0)  # replaces all missing values with 0
```

```python
data['Cabin'].fillna("Not Specified",inplace=True)
```

```python
data  # by using fillna() on series with 'inplace' we can make changes to dataframe
```

```python
data['Age'].head(10)
```

```python
data['Age'].fillna(method='bfill').head(10) # fills the next value in place of Nan
#35.0,Nan,54.0,2.0 => 35.0,54.0,54.0,2.0 
```

```python
data['Age'].fillna(method='ffill').head(10) # fills the prev. value in place of Nan
#35.0,Nan,54.0,2.0 => 35.0,35.0,54.0,2.0 
```

```python
series = data['Age'].fillna(0)
```
