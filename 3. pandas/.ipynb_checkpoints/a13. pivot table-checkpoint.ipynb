```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
# create a pivot table:
# rows - overs 1 to 20
# cols - all teams
# values - no. of 6's in that over in all seasons
```

```python
mask = delivery['batsman_runs'] == 6
six = delivery[mask]
six.head()
```

```python
six.shape
```

```python
# dataframe.pivot_table()
# index-> takes a list of rows for the table
# columns -> takes in which column values to use from dataframe
# values -> it takes in the column to compute values for the table
# aggfunc -> what operation to do with the 'values' parameter
# 'count' - counts the occurences in row
# 'mean' - takes the mean of all values corresponding to rows
```

```python
pt=six.pivot_table(index = ['over'], columns='batting_team', values='batsman_runs',aggfunc='count',fill_value=0)
```
