```python
# against which 3 teams did virat kohli score max. runs
```

```python
mask = delivery['batsman'] == "V Kohli"
```

```python
del_batsman = delivery[mask]
```

```python
runs_vkohli = del_batsman.groupby('bowling_team')
```

```python
runs_vkohli['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# we can define a function: 
```

```python
def runs_scored (batsman_name):
    del_batsman = delivery[delivery['batsman'] == batsman_name]
    return del_batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
runs_scored('V Kohli')
```
