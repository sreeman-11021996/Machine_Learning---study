```python
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
# x -> categorical data
# y -> numerical data
```

## a. boxplot() 

```python
sns.boxplot(x='total_bill',data=data)
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='box')
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='box',hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=data,hue='sex')
```

```python
# x -> categorical data
# y -> numerical data
```

## b. Violinplot()

```python
sns.violinplot(x = 'total_bill',data=data)
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='violin',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='violin',hue='sex',
            split=True)
```

```python
sns.violinplot(x='total_bill',y='day',data=data,hue='sex',split=True,
              orient='h')
```
