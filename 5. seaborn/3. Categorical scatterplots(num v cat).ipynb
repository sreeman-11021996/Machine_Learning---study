```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

## catplot() - figure level func

#### a. stripe - plot 

```python
# x -> categorical data
# y -> numerical data
```

```python
sns.catplot(x='day',y='tip',data=data,kind='strip')
```

```python
# jitter will introduce noise to displace the points with same values
```

```python
sns.catplot(x='day',y='tip',data=data,jitter=0,kind='strip')
```

```python
# hue semantic
```

```python
sns.catplot(x='day',y='tip',data=data,kind='strip',hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=data)
```

#### b. swarm-plot

```python
sns.catplot(x='day',y='tip',data=data,kind='swarm')
```

```python
sns.swarmplot(x='day',y='tip',data=data)
```
