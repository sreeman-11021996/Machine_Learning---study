```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

## a. barplot ()

```python
# x -> categorical values
# y -> numerical values
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar')
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex',
           estimator=np.median)
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex',
           estimator=np.var,row='day',col='time')
```

## b. countplot()

```python
sns.catplot(x='smoker',data=data,kind='count')
```

```python
sns.catplot(x='smoker',data=data,kind='count',hue='sex',col='day')
```

## pointplot() 

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='point')
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='point',col='sex',
           hue='day')
```

```python
sns.pointplot()
```
