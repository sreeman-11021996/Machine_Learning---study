```python
g = sns.FacetGrid(data,col='island',row='species',hue = 'sex',sharey=False,ylim=(10,25),
                 sharex=False,xlim=(30,50))
g.map_dataframe(sns.scatterplot,x='bill_length_mm',y='bill_depth_mm')
g.set_axis_labels(x_var='bill_length_mm',y_var='bill_depth_mm')
g.set_titles(col_template='{col_name} Island',row_template='{row_name}')
g.savefig
```
