## scatter => relplot() and scatterplot() 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# include a style of graphs to improve presentation
```

```python
plt.style.use('fivethirtyeight')
```

```python
# load a toy data set in seaborn library
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

## relplot()

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

#### hues parameter:

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

#### style parameter 

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

####  size parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',size='size',style='sex',sizes=(15,250),data=data)
```

## scatterplot() 

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',size='size',style='sex',sizes=(15,100),data=data)
plt.legend(bbox_to_anchor=(1.0, 1.25), borderaxespad=0)
```
