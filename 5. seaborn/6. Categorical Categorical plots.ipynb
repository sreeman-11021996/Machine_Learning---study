```python
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('flights')
```

```python
data.head()
```

```python
x=data.pivot_table(index='year',columns='month',values='passengers',
                 aggfunc='sum')
x
```

## a. heatmap()

```python
sns.heatmap(x,cmap='viridis')
```

```python
sns.heatmap(x,cbar=False,annot=True,fmt='d',linewidth=1)
```

## b. clustermap()

```python
# it clusters rows and cols based on dist. b/w the points
# mny typs of dist. -> euclid,chebyshev's,correlation....
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,z_score=0)
```

```python
sns.clustermap(x,z_score=0,row_cluster=False,metric='correlation')
```
